```python
import matplotlib.pyplot as plt
```

```python
def f(x:list[float]) -> list[float]:
    """
    Take square of a list\n
    x: a list of float of any values\n
    return: a list of square of input
    """
    return [i*i for i in x]
```

```python
xg = []
x0 = -10
lr = 0.1

for i in range(50):
    xg.append(x0)
    x0 -= lr*2*x0

```

```python
x = list(range(-10,11,1))

plt.plot(x,f(x), color='r')
plt.plot(xg, f(xg), 'bo')
plt.axline((0,0),(0,1))
plt.axline((0,0),(1,0))
plt.show()
```
